```python
#histogram variabel numerik
plt.hist(df['Release_date'], bins=20)
plt.xlabel('Release_date')
plt.ylabel('Tags')
plt.title('Price')
plt.show()

#bar plot untuk variabel kategorikal
sns.countplot(df['Review_no'])
plt.title('Review_type')
plt.show()
```

```python
#
plt.scatter(df['Review_no'], df['Review_type'])
plt.xlabel('Review_no')
plt.ylabel('Review_type')
plt.title('Review_no vs. Review_type')
plt.show()

#
sns.boxplot(x='Review_no', y='Review_type', data=df)
plt.title('Review_type by Review_no')
plt.show()
```

```python
#
correlation_matrix=df.corr()
sns.heatmap(correlation_matrix,annot=True,cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()

#
sns.pairplot(df[['Review_no', 'Review_type', 'Description']])
plt.title('Pair Plot')
plt.show()
```
